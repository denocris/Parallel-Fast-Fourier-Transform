```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata = np.loadtxt("data/timing_myfft2.txt")
fftwdata = np.loadtxt("data/timing_fftw2.txt")

n_procs = mydata[:,0]
myspeedup = mydata[0,1] / mydata[:,1]
fftwspeedup = fftwdata[0,1] / fftwdata[:,1]
```

```python
plt.title("Timing FFT: My Version vs FFTW Library")
plt.plot(n_procs[1::], mydata[1::,1],"-o", color="green", label="my_fft")
plt.plot(n_procs[1::], fftwdata[1::,1], "-o", color="red", label="fftw")
plt.ylabel("time (s)")
plt.xlabel("num procs")
plt.legend(loc="upper left")
plt.annotate('Size = 512', xy=(0.30, 0.9), xycoords='axes fraction')
plt.savefig("timing_size512.png")
plt.show()
```

```python

plt.title("Speedup FFT: My Version vs FFTW Library")
plt.plot(n_procs, n_procs, "-o", color="blue")
plt.plot(n_procs, myspeedup,"-o", color="green", label="my_fft")
plt.plot(n_procs, fftwspeedup, "-o", color="red", label="fftw")
plt.ylabel("Speedup")
plt.xlabel("num procs")
plt.legend(loc="upper left")
plt.annotate('Size = 512', xy=(0.30, 0.9), xycoords='axes fraction')
plt.savefig("Speedup_size512.png")
plt.show()
```
